```
import requests
from bs4 import BeautifulSoup
```

```
# HTML 코드 받아오기
response = requests.get("https://workey.codeit.kr/ratings/index")
rating_page = response.text

# BeautifulSoup 사용해서 HTML 코드 정리

soup = BeautifulSoup(rating_page, 'html.parser')
print(soup.select('table'))
```

```
print(soup.select('td.program'))
```

```
# "program" 클래스를 가진 td 태그 선택하기

program_title_tags = soup.select('td.program')

program_titles = []

# 태그 안에 있는 text를 가져오기 위해서는 get_text() method 쓸 수 있음.

for tag in program_title_tags :
    program_titles.append(tag.get_text())
    print(tag.get_text())

print(program_titles)
```

```
program_title_tags = soup.select('td.program')

# 가장 위에 있는 내용 뽑아줌.

print(soup.select_one('td.program'))
```

```
response = requests.get("https://workey.codeit.kr/ratings/index")
rating_page = response.text

soup = BeautifulSoup(rating_page, 'html.parser')

td_tags = soup.select('td')[:4] # td 태그 중 앞에서 4개까지만 자른다.

for tag in td_tags:
    print(tag.get_text())
```
